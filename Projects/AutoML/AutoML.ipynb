### Введение
Мотивация: Современные модели глубокого обучения (DL) добились выдающихся результатов в задачах компьютерного зрения, обработки языка и других областях. Однако их создание требует значительных усилий экспертов.

AutoML позволяет автоматизировать весь ML-пайплайн, позволяя даже неспециалистам строить качественные модели без глубоких знаний в области ИИ.

Автоматизация полного пайплайна машинного обучения включает в себя следующие этапы:

- Подготовка данных

- Инженерия признаков

- Генерация модели

- Настройка гиперпараметров

- Оценка модели

### Подготовка данных

Три основных этапа:

1. Сбор данных:

  - Использование готовых датасетов (Kaggle, ImageNet)

  - Преобразование данных из интернета (с фильтрацией и самоподписыванием)

  - Генерация синтетических данных с помощью симуляторов и GAN.

2. Очистка данных:

  - Отбор и фильтрация шумных и ошибочных данных (вручную).

  ИЛИ

  - Автоматизированная очистка:

    - BoostClean: выбирает набор очисток с помощью бустинга.

    - AlphaClean: трансформирует очистку в задачу гиперпараметров.

    - Gemp et al. — применение метаобучения.



3. Аугментация данных:

(Автоматизированный процесс искусственного увеличения объема и разнообразия обучающих данных с помощью различных преобразований или генерации синтетических примеров. Это помогает улучшить качество моделей машинного обучения, особенно когда исходный набор данных мал или несбалансирован.)

  - Изображения:

    - Аффинные: повороты, обрезки.

    - Эластичные: яркость, размытие.

    - Нейросетевые: GAN, стиль, adversarial noise.

  - Текст:

    - Замены синонимов, переводы туда-обратно.

    - NLPAug — библиотека аугментаций.

  - Автоматический подбор:

    - AutoAugment — поиск политики RL.

    - Более быстрые методы: RandAugment, FastAugment, UniformAugment.

### Инженерия признаков
Подзадачи:

1. Отбор признаков: выбор значимых признаков (фильтры, обёртки, встроенные методы).

2. Построение признаков: создание новых признаков путём преобразования или комбинаций.

3. Извлечение признаков: уменьшение размерности (PCA, автоэнкодеры, нейросети).

### Генерация моделей

Подразделяется на:

- Пространства поиска архитектур (определяет, какие архитектуры моделей могут быть рассмотрены):

  - Целостные (entire-structured):
  
    Каждая операция (слой) выбирается независимо (последовательно).

  - Ячеистые (cell-based):
  
    Архитектура состоит из повторяющихся "ячеек" (cells). AutoML ищет оптимальную структуру одной ячейки, которая затем тиражируется.

  - Иерархические (hierarchical):
  
    Иерархическая организация ячеек, где высокоуровневые ячейки состоят из низкоуровневых (PNAS, MnasNet).

  - Морфизм-базированные (morphism-based):
  
    Основаны на трансформациях существующих сетей (Net2Net, Network Morphism).





- Методы архитектурной оптимизации (описывают, как искать лучшую модель в заданном пространстве):

  - Эволюционные алгоритмы (EA): основаны на отборе, мутации, скрещивании.

  - Обучение с подкреплением (RL): контроллер RNN, генерирующий архитектуры.

  - Градиентные методы (GD): DARTS, непрерывное представление архитектур.

  - Суррогатное моделирование (SMBO): предсказание качества архитектур с помощью BO, RF или нейросетей.

  - Простые методы (Grid/Random search).

  - Гибридные методы: комбинации EA, RL, GD и SMBO.

- Оценка моделей:

  - Полное обучение модели — дорого.

  - Упрощённые методы:

    - Low-fidelity - уменьшение эпох/размера.

    - Early-stopping - ранняя остановка.

    - Weight sharing - параметрическое разделение весов.

    - Surrogate - суррогатные модели.


NAS (Neural Architecture Search) -  это автоматический подбор архитектуры модели для конкретной задачи: какие слои, в каком порядке, с какими связями и размерами.

1. Определяется пространство архитектур:
Какие комбинации и блоки можно использовать.

2. Алгоритм поиска (например, reinforcement learning или эволюционные алгоритмы) перебирает архитектуры:

- строит модель,

- обучает (полностью или частично),

- измеряет качество.

3. Сохраняется лучшая архитектура.

# AutoGluon-Tabular

AutoGluon-Tabular — это открытый AutoML-фреймворк от Amazon, который позволяет буквально в одной строке кода обучить мощную модель на табличных данных без предварительной подготовки. Он работает с данными таблиц, изображениями и текстом, способен выполнять задачи классификации, регрессии, а также более сложные задания. Autogluon позволяет автоматизировать процесс построения моделей машинного обучения.

Что делает (автоматически) AutoGluon кратко:

- определяет тип задачи (регрессия / классификация);

- обрабатывает пропуски и категории;

- подбирает и обучает множество моделей;

- объединяет их в ансамбль (ensemble / stacking);

- находит лучшую комбинацию и возвращает финальные предсказания.

Главные особенности:

- Пользователь может задавать лимит по времени обучения. (За отведенное время AutoGluon подыщет наиболее эффективную модель)

- Поддержка "сырых" данных. Нет нужды в предварительной обработке или ручном указании типов признаков.

- Автоматическое распознавание задачи. AutoGluon сам определяет, классификация это или регрессия.

- Продвинутое ансамблирование. Вместо выбора одной лучшей модели, AutoGluon использует многоуровневый стекинг и повторное k-fold bagging.

- Есть возможность прерывать и продолжать обучение

### API fit и обучение модели
Рассмотрим минимальный пример:

```python
from autogluon.tabular import TabularPredictor

predictor = TabularPredictor(label='price').fit("train.csv")
predictions = predictor.predict("test.csv")
```

Внутри fit() происходит:

- Инференс задачи (классификация/регрессия),

- Обучение множества моделей,

- Стекинг,

- Сохранение результатов на диск (для возможности продолжения обучения).

Также можно добавлять различные параметры.

```python
fit(..., time_limit=3600, hyperparameter_tune=True, auto_stack=True, eval_metric='mape')
```

### Предобработка данных

AutoGluon сначала проводит model-agnostic preprocessing:

- Определяет тип признаков: числовые, категориальные, текстовые, временные.

- Текст превращается в n-граммы.

- Даты разбиваются на год, месяц и т.п.

- Пропущенные значения категориальных признаков заменяются на Unknown.

Далее — model-specific preprocessing, уникальный для каждой модели:

- Для деревьев: one-hot или ordinal encoding.

- Для нейросетей: embedding + нормализация чисел

### Нейросетевые модели в AutoGluon

AutoGluon использует гибкую tabular neural network, которая разрабатывалась специально под табличные данные и показала хорошее обобщение при обучении в ансамблях.

Особенности архитектуры:

- Input block: обрабатывает числовые, категориальные и текстовые признаки.

  - Категориальные переменные → через эмбеддинги (преобразование данных в числовые векторы).

  - Текстовые — через n-граммы и simple pooling.

- Residual connections (skip-соединения): применяются между слоями, чтобы избежать переобучения.

- Normalization layers: применяются к каждому признаку отдельно (per-feature normalization).

- Dropout и batch normalization: для регуляризации.

- Final block: линейный слой с софтмаксом (для классификации) или регрессией.


### Бэггинг в AutoGluon:

Бэггинг, или Bootstrap Aggregation, является техникой, используемой в машинном обучении для улучшения стабильности и точности алгоритмов. Основная идея заключается в том, что комбинация нескольких моделей обычно приводит к лучшей производительности, чем любая отдельная модель. Это связано с тем, что бэггинг помогает уменьшить переобучение и повысить надежность предсказаний.

- Разделение данных: Обучающие данные делятся на K частей (или фолдов).

- Обучение моделей: Для каждого из фолдов обучается модель, используя все данные, кроме этого фолда. Это дает возможность создавать K экземпляров модели с различными частями данных, что называется bagged моделью.

- Кросс-валидация: Каждый экземпляр модели оценивается на основе фолда, который не использовался для обучения. Полученные предсказания из всех фолдов объединяются, чтобы создать предсказания вне фолда (OOF). Эти предсказания играют критическую роль в процессе стекинга, поскольку они позволяют избежать утечки данных.

- Агрегация: При предсказании бэггинг объединяет все индивидуальные модели и усредняет их предсказания для генерации финального ответа.

### Стекинг в AutoGluon:

- Слои моделей: Стекинг напоминает многоуровневый торт. Каждый слой состоит из нескольких различных bagged моделей, использующих предсказания из предыдущего слоя в качестве входных признаков, наряду с оригинальными признаками из обучающих данных.

- Взвешенный ансамбль: Последний слой содержит одну "супер" модель, которая объединяет предсказания из предпоследнего слоя. Эта модель, известная как meta-model, обучается для объединения выходов всех предыдущих моделей и делает финальное предсказание.

- Обработка утечки данных: Во время обучения важно избегать утечки данных, поэтому используются предсказания вне фолда (OOF) каждой bagged модели, вместо того, чтобы предсказывать непосредственно по обучающим данным.

### Итог:

Таким образом, AutoGluon строит stacked ensembles с несколькими уровнями (multi-layer stacking) и повторным бэггингом. Это ключ к высокой точности.

Компоненты стекинга:

- Level 0 — базовые модели (CatBoost, LightGBM, XGBoost, Random Forest, KNN, Neural Net).

- Level 1, 2 и т.д. — модели, обученные на предсказаниях предыдущих уровней.

Повторный k-fold bagging:

- Каждая модель обучается на разных подвыборках с перекрестной валидацией.

- Улучшает устойчивость к переобучению и повышает обобщающую способность.

AutoGluon умеет автоматически управлять временем обучения, чтобы не выйти за пределы time_limit.

- Строится несколько моделей.

- Замеряется среднее время на одну модель.

- Вычисляется, сколько моделей реально успеть обучить.

- Распределяется бюджет между моделями и стекинг-уровнями.

### Настройка гиперпараметров

AutoGluon поддерживает:

### Пример использования в задаче с недвижимостью

```python
from autogluon.tabular import TabularPredictor
import pandas as pd

# === Загрузка данных ===
train_data = pd.read_csv("train.csv")
test_data = pd.read_csv("test.csv")

# Целевая переменная
label = 'price'

# === Обучение модели ===
predictor = TabularPredictor(
    label=label,
    eval_metric='mape',  # метрика оценки
    verbosity=2
).fit(
    train_data=train_data,
    time_limit=1800,  # максимум на обучение
    presets='best_quality'  # или 'medium_quality_faster_train' — быстрее
)

# === Предсказание ===
preds = predictor.predict(test_data)

# === Подготовка submission ===
submission = pd.DataFrame({
    "id": test_data["Unnamed: 0"] if "Unnamed: 0" in test_data.columns else test_data["id"],
    "price": preds
})

submission.to_csv("submission.csv", index=False)
print("Файл submission.csv готов!")
```

Тонкая настройка обучения:

```python
from autogluon.tabular import TabularPredictor
import pandas as pd

# === Загрузка данных ===
train_data = pd.read_csv("train.csv")
test_data = pd.read_csv("test.csv")

# Целевая переменная
label = 'price'

# === Обучение модели ===
predictor = TabularPredictor(
    label='price',
    eval_metric='mape'
).fit(
    train_data,
    time_limit=1800,
    presets='medium_quality_faster_train',
    num_bag_folds=5,
    num_stack_levels=1,
    ag_args_fit={"num_gpus": 1}
)

# === Предсказание ===
preds = predictor.predict(test_data)

# === Подготовка submission ===
submission = pd.DataFrame({
    "id": test_data["Unnamed: 0"] if "Unnamed: 0" in test_data.columns else test_data["id"],
    "price": preds
})

submission.to_csv("submission.csv", index=False)
print("Файл submission.csv готов!")
```
