```python
from transformers import MambaConfig, MambaForCausalLM, AutoTokenizer
import torch

tokenizer = AutoTokenizer.from_pretrained("state-spaces/mamba-370m-hf")
model = MambaForCausalLM.from_pretrained("state-spaces/mamba-370m-hf")
```

```python
prompt = "Продолжи: Я помню чудное мгновенье: Передо мной явилась ты, Как мимолетное виденье, Как гений чистой "
inputs = tokenizer(prompt, return_tensors="pt")

# Generate
generate_ids = model.generate(inputs.input_ids, max_length=250)
tokenizer.batch_decode(generate_ids, skip_special_tokens=True, clean_up_tokenization_spaces=False)[0]
```

```python
prompt = """I loved you, and that love to die refusing,
May still - who knows! Be smoldering in my breast.
Pray, be not pained -- believe me, of my choosing
I'd never have you troubled nor yet distressed.
I loved you mutely, hopelessly and truly,
With shy yet fervent, tenderness aglow;
Mine was a jealous passion and unruly….
May Heaven grant another loves you so!
Write the same lyrics as above, starting with this:
A monument I've raised not built with hands,
"""
inputs = tokenizer(prompt, return_tensors="pt")

# Generate
generate_ids = model.generate(inputs.input_ids, max_length=len(prompt) + 200)
tokenizer.batch_decode(generate_ids, skip_special_tokens=True, clean_up_tokenization_spaces=False)[0]
```
